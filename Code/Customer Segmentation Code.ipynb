Importing libraries and data set

```python
import pandas as pd
import numpy as np

dataset = pd.read_csv("../3. Uploaded Data/20201030/Mall_Customers.csv")
dataset.head()
```

Create X data set 

```python
X_set = dataset.iloc[: ,3:].values
```

Using K-Mean algorithm to cluster our cutomers

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters =5,init = 'k-means++', max_iter = 300, n_init =10,random_state=21)
y_set = kmeans.fit_predict(X_set)
```

```python
print(y_set)
```

Visualizing customer clusters

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,15))
plt.scatter(X_set[y_set==0, 0], X_set[y_set==0, 1], c='red',label='Budget Buyer')
plt.scatter(X_set[y_set==1, 0], X_set[y_set==1, 1], c='blue',label='Miser')
plt.scatter(X_set[y_set==2, 0], X_set[y_set==2, 1], c='orange',label='"Golden Youth"')
plt.scatter(X_set[y_set==3, 0], X_set[y_set==3, 1], c='cyan',label='Laborer')
plt.scatter(X_set[y_set==4, 0], X_set[y_set==4, 1], c='green',label='Spender')

plt.title('Clusters of clients')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending score (1-100)')
plt.legend()
plt.show()
```
